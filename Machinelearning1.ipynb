ENCODING THE CATEGORICAL VARIABLES 

```python
cols=["Restaurant Name","City","Address","Locality","Locality Verbose","Cuisines","Average Cost for two","Currency","Has Table booking","Has Online delivery","Is delivering now","Switch to order menu","Rating color","Rating text"]
le=LabelEncoder()
d=df.copy()
for i in cols:
 df[i]=le.fit_transform(df[i])
 print(df)
```

Let us now seperate the data into predictors and response 

```python
df.columns
```

```python
X=df.drop(columns="Cuisines",axis=1)
Y=df["Cuisines"]
```

Now let us split the data into training data and test data 

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=.1,random_state=42)
```

```python
print(X.shape)
print(X_train.shape)
print(X_test.shape)
```

```python
scaler=StandardScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.fit_transform(X_test)
```

We see that the data has already been divided into training data and test data

MODEL CREATION AND TRAINING THE MODEL USING TRAINED DATA

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
lr = RandomForestClassifier(n_estimators=200,max_depth=15,min_samples_split=5,min_samples_leaf=2,class_weight='balanced',random_state=42,n_jobs=1)
lr.fit(X_train, Y_train)
```

```python
Y_predicted=lr.predict(X_train)
```

ACCURACY SCORES

```python
accuracy=accuracy_score(Y_train,Y_predicted)
accuracy
```

```python
Y_predictedtest=lr.predict(X_test)
```

```python
accuracy_test=accuracy_score(Y_test,Y_predictedtest)
accuracy_test
```

```python
from sklearn.metrics import precision_score
precision=precision_score(Y_test,Y_predictedtest,average="macro")
```
