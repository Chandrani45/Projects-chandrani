```python
import os,sys#operating system,sys is system
sys.executable
```

```python
os.environ["PYSPARK_PYTHON"]=sys.executable
```

```python
import pyspark
sc=pyspark.SparkContext(master="local[*]",appName="FirstExample")#It uses all the core of the cpu
sc.setLogLevel("ERROR")#To display all the error
```

```python
print(sc)#It is like a pool processor executor
```

```python
#Create your first RDD
data=list(range(8))
rdd=sc.parallelize(data)#create collection
rdd.collect()
```

```python
rdd.isEmpty()
```

```python
mixRDD=sc.parallelize([True,[11,22,33,44,55],(10,20,30,40,50)],3)
mixRDD.collect()
```
